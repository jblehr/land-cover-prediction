```python
import optuna
import convGRU
from optuna.trial import TrialState

name = 'initial_importance_test'
# storage = f"sqlite:////content/drive/Shareddrives/Adv-ML Project/repo/land-cover-prediction//output/{name}.db"
storage = f"sqlite:////media/nick/Data/GitRepos/land-cover-prediction/output/hail_mary_local.db"

study = optuna.create_study(direction="minimize", study_name=name, storage=storage, load_if_exists=True)
study.optimize(convGRU.objective, n_trials=100)

pruned_trials = study.get_trials(deepcopy=False, states=[TrialState.PRUNED])
complete_trials = study.get_trials(deepcopy=False, states=[TrialState.COMPLETE])

print("Study statistics: ")
print("  Number of finished trials: ", len(study.trials))
print("  Number of pruned trials: ", len(pruned_trials))
print("  Number of complete trials: ", len(complete_trials))

print("Best trial:")
trial = study.best_trial

print("  Value: ", trial.value)

print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))

fig = optuna.visualization.plot_param_importances(study)
fig.show()
```
