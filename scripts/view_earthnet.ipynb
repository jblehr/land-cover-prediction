```python
fig = en.cube_gallery(cubepath, variable='ndvi')
# https://github.com/earthnet2021/earthnet-toolkit/blob/master/earthnet/plot_cube.py
# variable (str, optional):  One of "rgb", "ndvi", "rr","pp","tg","tn","tx". Defaults to "rgb".
# rr: precip
# pp: pressure
# tg: mean temp
# tn: min temp
# tx: max temp
plt.show()

# Shows 30 images (1 every 5 days) for rgb and ndvi; 150 images (1 every day) for metereological variables
```
